## Image Matching Example

```python
from src.ImageMatching import ImageMatching
```

```python
base_path =  r"outputs/nothing - bruno major/nothing - bruno major_20250119_164547_unique/"
image_numbers = [82, 85]
image_paths = [f"screenshot_{num:04d}.png" for num in image_numbers]
print(image_paths)
image_paths = [base_path + path for path in image_paths]
```

```python
SET_OVERRIDE = 120
TOP_N = 0
MANUAL_OVERRIDE = [475, 27]
matcher = ImageMatching(image_paths, overlap_region=SET_OVERRIDE,
                        use_top_n=TOP_N, manual_override=MANUAL_OVERRIDE)

matcher.plot_images()
```

```python
matcher.plot_feature_matching(do_print=False)
```

```python
matcher.plot_stitched_image()
```
